```python
from PIL import Image
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import jieba
import numpy as np
from collections import Counter
```

```python
text = open('D:\大學作業\三年級\人工智慧\斷詞與文字雲教學/joker.txt',"r",encoding="utf-8").read() #讀文字資料

jieba.set_dictionary('D:\大學作業\三年級\人工智慧\斷詞與文字雲教學/chinese.txt')
with open('D:\大學作業\三年級\人工智慧\斷詞與文字雲教學/stop.txt','r',encoding='utf-8-sig') as f:
    stops = f.read().split('\n')
terms = [] #儲存字詞
for t in jieba.cut(text,cut_all=False): #拆解句子為字詞
    if t not in stops: #不是停用詞
        terms.append(t)
diction = Counter(terms)
# 可列出詞的統計數量
#print(diction)

font = "C:\Windows\Fonts\kaiu.ttf" #設定字體(標楷體)

mask = np.array(Image.open("D:\大學作業\三年級\人工智慧\斷詞與文字雲教學\jokersmile.jpg")) #設定文字雲形狀

wordcloud = WordCloud(background_color="white",mask=mask, font_path=font) #背景顏色預設黑色,改為白色,字體改為標楷體
wordcloud.generate_from_frequencies(diction) #產生文字雲

#產生圖片
plt.figure(figsize=(6,6))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()

wordcloud.to_file("news worldcloud.png") #存檔
```
